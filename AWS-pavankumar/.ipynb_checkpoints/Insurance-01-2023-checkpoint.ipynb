```python
def find_unique(data,cols):
        
    print('Returns a DataFrame that contains information about unique values for the column names provided.')

    cnull=[sum(data[y].isnull()) for y in data.columns]
    uq=pd.DataFrame({'Col_name':cols,
                     'Count_':list(len(data[x].unique()) for x in cols),
                    'Unique':list(data[x].unique() for x in cols),
                      })
    return uq
```

```python
# Preparing a function that returns information on missing values

def check_miss(data):
    cnull=[sum(data[y].isnull()) for y in data.columns]
    miss=pd.DataFrame({'Col_name':data.columns,'Null Values':
                        [any(data[x].isnull()) for x in data.columns],
                    'Count_Nulls':cnull,
                    'Percentage_Nulls':list((np.array(cnull)*100)/data.shape[0]),
                    'DataType':list(data[x].dtype for x in data.columns)
                      })
    return miss.sort_values(by='Count_Nulls',ascending=False)
```

```python
miss_vals = check_miss(df)
miss_vals
```

 Returns a sorted DataFrame that contains information about missing values in each column.
 
 Null Values: Either the column has null values or not. 
              As we can see smoker and sex have null values.
        
 Count_Nulls: Number of null values.
 
 Percentage_Nulls: Percentage of null values

In above cell we found missing values in categorical features so, we will try to replace with mode value.

```python
df2 = miss_vals[miss_vals['Count_Nulls'].values > 0]

def change_null(df):
    for val in df2['Col_name']:
        df[val].fillna(df[val].mode()[0],inplace=True)
```

```python
# outlier treatment:
def outlier_treatment(datacolumn):
    sorted(datacolumn)
    q1,q3 = np.percentile(datacolumn,[25,75])
    iqr = q3 -q1
    lower_range = q1 - (1.5 * iqr)
    upper_range = q3 + (1.5 * iqr)
    return lower_range,upper_range,iqr
```

```python
l1,u1,iqr1 = outlier_treatment(df['age'])
l2,u2,iqr2 = outlier_treatment(df['bmi'])
print(l1,u1,iqr1,l2,u2,iqr2)
```

```python
# Preparing a function that returns information on missing values

def outlier_data(cols):
    count = []
    for col in cols:
        if col == numeric_features[0]:
            count.append(df[(df[col] < l1) | (df[col] > u1)].shape[0])
        else:
            count.append(df[(df[col] < l2) | (df[col] > u2)].shape[0])
    
    miss1=pd.DataFrame({'Col_name':numeric_features[:-1],
                       'Lower_bound':[l1,l2],
                       'Upper_bound':[u1,u2],
                       'iqr': [iqr1,iqr2],
                    'No of outliers':count,
                    'Percentage_outliers':list((np.array(count)*100)/df.shape[0]),
                      })
    return miss1.sort_values(by='No of outliers',ascending=False)
```

```python
df3 = outlier_data(numeric_features[:-1])
df3
```

Here we can see outlier's data  

```python
for val in numeric_features[:-1]:
    lr,ur,iqr = outlier_treatment(df[val])
    print(l1,u1)
    df[val] = np.where(df[val] > ur, ur,df[val])
    df[val] = np.where(df[val] < lr, lr,df[val])
            
```

```python
#Multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
calc_vif(df[[i for i in df.describe().columns if i not in ['charges']]])
```

```python
for col in numeric_features[:-1]:
    fig=plt.figure(figsize=(9,6))
    ax=fig.gca()
    mean = 'mean' + ' : '+ str(round(df[col].mean(),3))
    median = 'median'+' : '+ str(round(df[col].median(),3))
    print('skewness of ' + col +" : " + str(skew(df[col])))
    sns.distplot(df[col])
    ax.axvline(df[col].mean(),color='magenta', linestyle='dashed',linewidth=4)
    ax.axvline(df[col].median(), color='cyan', linestyle='dashed', linewidth=4)
    plt.legend([mean,median])
    ax.set_title(col)
plt.show()

print('If skewness < 0: Left skewed or Negatively skewed' + '\n' + "If skewness >0: Right skewed or Positive skewed"  + 'If skewness =0: Normally Distributed')
```
